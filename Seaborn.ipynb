# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

- Scatter Plot:
 Uses: To visualize the relationship between two numerical variables. It can show patterns, clusters, or correlations between variables.
 
- Line Plot:
Uses: To visualize the trend or evolution of a numerical variable over time or any other continuous variable.

- Bar Plot:

 Uses: To compare the values of a categorical variable across different groups or categories. It can display the average, count, or any other summary statistic of the data.

- Box Plot:
 Uses: To visualize the distribution of a numerical variable across different categories. It shows the median, quartiles, and any outliers or extreme values in the data.
 
- Heatmap:
Uses: To display the relationship between two categorical variables by using color intensity. It is useful for identifying patterns, correlations, or associations in large datasets.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

```python
import seaborn as sns

# Load the "fmri" dataset
fmri = sns.load_dataset('fmri')

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns

# Load the "titanic" dataset
titanic = sns.load_dataset('titanic')

# Box plots
sns.boxplot(data=titanic, x='pclass', y='age')
sns.boxplot(data=titanic, x='pclass', y='fare')
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the "diamonds" dataset
diamonds = sns.load_dataset('diamonds')

# Histogram with hue
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
import seaborn as sns

# Load the "iris" dataset
iris = sns.load_dataset('iris')

# Pair plot with hue
sns.pairplot(data=iris, hue='species')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns

# Load the "flights" dataset
flights = sns.load_dataset('flights')

# Reshape the data for heatmap
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Heatmap
sns.heatmap(data=flights_pivot, cmap='Blues', annot=True, fmt='d')
```
