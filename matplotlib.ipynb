# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib

- Matplotlib is a data visualization library in Python used for creating various types of plots and charts. It provides a wide range of tools for creating static, animated, and interactive visualizations.

- Matplotlib is used to visualize data and present it in a meaningful way. It is often used for exploratory data analysis, communicating insights, and creating publication-quality figures.

- The Pyplot module of Matplotlib provides a simple interface for creating and customizing plots. Some of the common types of plots that can be created using Pyplot include:

- Line Plot: Shows the relationship between two variables by connecting the data points with lines.
- Bar Plot: Displays categorical data using rectangular bars with heights or lengths proportional to the values they represent.
- Scatter Plot: Shows the relationship between two variables as a collection of points.
- Histogram: Represents the distribution of a continuous variable by dividing it into bins and displaying the count or frequency of observations in each bin.
- Pie Chart: Represents categorical data as a circle divided into sectors, where each sector represents a specific category and its size corresponds to the proportion of data in that category.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Plot 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

# Adjusting spacing between subplots
plt.tight_layout()

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 4))
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 4))
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Data for box plots
data = [box1, box2]

# Box plot
plt.figure(figsize=(8, 4))
plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.show()
```
